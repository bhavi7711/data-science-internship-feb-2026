## Problem 1: Employee Performance Bonus Eligibility

### Description:
A company evaluates employee performance scores at the end of the year.

### Objective:
- Identify the highest performance score.
- Handle ties if multiple employees have the same highest score.
- Display all employees eligible for the top performance bonus.


```python
# Problem 1: Employee Performance Bonus Eligibility

# Step 1: Define employee performance data
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 2: Find the highest performance score
highest_score = max(employees.values())

# Step 3: Identify employees who scored the highest
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Step 4: Display result
print("Top Performers Eligible for Bonus:")
print(f"{', '.join(top_performers)} (Score: {highest_score})")
```

## Problem 2: Search Query Keyword Analysis

### Problem Description:
An e-commerce website stores customer search queries.

### Requirements:
- Convert input to lowercase.
- Remove punctuation.
- Count frequency of each keyword.
- Display only keywords searched more than once.


```python
# Problem 2: Search Query Keyword Analysis

import string

# Step 1: Define input query
query = "Buy mobile phone buy phone online"

# Step 2: Convert to lowercase
query = query.lower()

# Step 3: Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Step 4: Split sentence into words
words = query.split()

# Step 5: Count frequency of each word
frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Step 6: Filter words appearing more than once
repeated_words = {}

for word, count in frequency.items():
    if count > 1:
        repeated_words[word] = count

# Step 7: Display result
print("Keywords searched more than once:")
print(repeated_words)
```

## Problem 3: Sensor Data Validation

### Problem Description:
A factory collects sensor readings every hour.

### Requirements:
- Identify even readings (valid readings).
- Store them as (hour_index, reading_value) pairs.
- Ignore odd readings.


```python
# Problem 3: Sensor Data Validation

# Step 1: Define sensor readings
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Step 2: Create empty list to store valid readings
valid_readings = []

# Step 3: Loop through readings
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:  # Check if value is even
        valid_readings.append((index, value))

# Step 4: Display valid readings
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## Problem 4: Email Domain Usage Analysis

### Problem Description:
Analyze which email providers employees are using.

### Requirements:
- Count number of users per domain.
- Calculate percentage usage for each domain.


```python
# Problem 4: Email Domain Usage Analysis

# Step 1: Define email list
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Step 2: Count domain occurrences
domain_count = {}

for email in emails:
    domain = email.split("@")[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Step 3: Calculate percentage usage
total_users = len(emails)

print("Email Domain Usage Percentage:")

for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

## Problem 5: Sales Spike Detection

### Problem Description:
Detect days where sales are more than 30% above the average.

### Requirements:
- Calculate average daily sales.
- Identify spike days.


```python
# Problem 5: Sales Spike Detection

# Step 1: Define sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 2: Calculate average sales
average_sales = sum(sales) / len(sales)

# Step 3: Calculate threshold (30% above average)
threshold = average_sales * 1.3

# Step 4: Detect spike days
print("Sales Spike Days:")

for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

## Problem 6: Duplicate User ID Detection

### Problem Description:
Detect duplicate user IDs in a system.

### Requirements:
- Identify duplicate IDs.
- Display how many times each duplicate appears.


```python
# Problem 6: Duplicate User ID Detection

# Step 1: Define user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Step 2: Count occurrences
frequency = {}

for user in user_ids:
    if user in frequency:
        frequency[user] += 1
    else:
        frequency[user] = 1

# Step 3: Display duplicates
print("Duplicate User IDs:")

for user, count in frequency.items():
    if count > 1:
        print(f"{user} → {count} times")
```
