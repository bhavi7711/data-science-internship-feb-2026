## Problem 1: Social Media – Post Engagement Analyzer

This function calculates total likes and determines whether the post is viral.
If total likes ≥ 1000 → Viral Post
Else → Normal Engagement

```python
def post_engagement_analyzer(likes_list):
    total_likes = 0

    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

post_engagement_analyzer([200, 300, 400, 150])
```

## Problem 2: Healthcare – Medicine Stock Alert System

Checks whether medicine stock is low or sufficient.
Stock < 10 → Low Stock Alert
Stock ≥ 10 → Stock Sufficient

```python
def medicine_stock_alert(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)

medicine_stock_alert(6)
```

## Problem 3: Agriculture – Rainfall Adequacy Checker

Calculates average rainfall and checks adequacy.

```python
def rainfall_checker(rainfall_data, required_level):
    total = 0

    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

rainfall_checker([60, 75, 80], 70)
```

## Problem 4: Social Media – Duplicate Account Detection

Detects duplicate usernames using set.

```python
def duplicate_account_detection(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

duplicate_account_detection(["user1", "user2", "user1"])
```

## Problem 6: Agriculture – Premium Crop Price Filter

```python
def premium_crop_filter(prices):
    premium_crops = []

    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)

premium_crop_filter([1500, 2500, 3200, 1800])
```

## Problem 7: System Monitoring – Application Health Checker

```python
def application_health_checker(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

application_health_checker(7)
```

## Problem 9: E-Learning – Attendance Eligibility System

```python
def attendance_checker(attendance_list):
    total_classes = len(attendance_list)
    attended = 0

    for status in attendance_list:
        if status == "Present":
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)

attendance_checker(["Present", "Present", "Absent", "Present", "Present"])
```

## Problem 10: Smart Electricity Bill Analyzer

```python
def electricity_bill_analyzer(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Classification:", usage)

electricity_bill_analyzer(250)
```
