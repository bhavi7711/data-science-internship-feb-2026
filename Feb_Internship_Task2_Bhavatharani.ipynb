# Problem 1: Unique Words in a Sentence

Objective:
To identify and display all unique words in a given sentence along with their count.

Approach:
1. Take input from the user.
2. Split the sentence into words.
3. Convert the list into a set to remove duplicates.
4. Display unique words and their count.


```python
# Problem 1: Unique Words in a Sentence

# Taking input from the user
sentence = input("Enter sentence: ")

# Splitting sentence into individual words
words = sentence.split()

# Converting list into set to remove duplicates
unique_words = set(words)

# Displaying results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

# Problem 2: Highest Salary from Employee Data

Objective:
To determine which employee has the highest salary from a dictionary.

Approach:
1. Store employee names and salaries in a dictionary.
2. Use max() to find the employee with highest salary.
3. Display the result.


```python
# Problem 2: Highest Salary from Employee Data

# Dictionary of employees and salaries
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Finding employee with highest salary
highest_employee = max(employees, key=employees.get)

# Displaying result
print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

# Problem 3: Find Maximum and Minimum Values

Objective:
To find the maximum and minimum values in a list of numbers.

Approach:
1. Initialize max and min with first element.
2. Traverse through list using loop.
3. Compare and update values accordingly.


```python
# Problem 3: Find Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]

# Initializing max and min
maximum = numbers[0]
minimum = numbers[0]

# Loop to compare values
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Displaying results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

# Problem 4: Count Products Above a Price Threshold

Objective:
To count the number of products priced above 1000.

Approach:
1. Store prices in a list.
2. Use loop to compare each price with 1000.
3. Count matching items.


```python
# Problem 4: Count Products Above 1000

prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

# Problem 5: Calculate Attendance Percentage

Objective:
To calculate attendance percentage from a list of 'P' and 'A'.

Approach:
1. Count number of 'P'.
2. Divide by total number of days.
3. Multiply by 100.


```python
# Problem 5: Calculate Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P")
total_days = len(attendance)

percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```
