## Problem 1: Mobile Recharge Validation System

### Rules:
- Recharge amount must be ≥ ₹50
- Valid plans: 199, 299, 399, 599
- Allow retry if invalid

### Real-Time Use:
- Telecom recharge apps
- Payment gateways


```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False

    if amount not in valid_plans:
        print("Invalid recharge plan selected.")
        return False

    print("Recharge successful!")
    return True


# Retry using while loop
while True:
    try:
        user_amount = int(input("Enter recharge amount: "))
        if validate_recharge(user_amount):
            break
    except ValueError:
        print("Please enter a valid number.")
```

## Problem 2: Inventory Reorder Alert System

### Rules:
- Print "Reorder Alert" if stock < 15
- Print "Stock OK" otherwise

### Real-Time Use:
- Warehouse systems
- Retail management


```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")


inventory = {
    "Rice": 10,
    "Wheat": 25,
    "Oil": 12,
    "Sugar": 30
}

check_inventory(inventory)
```

## Problem 3: Student Result Processing System

### Rules:
- Accept list of marks
- Calculate average
- Pass if average ≥ 50
- Fail otherwise

### Real-Time Use:
- Online exam systems
- Evaluation portals


```python
# Problem 3: Student Result Processing System

def process_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    print(f"Average Marks: {average}")

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


student_marks = [60, 55, 40, 70, 50]

result = process_result(student_marks)
print("Result:", result)
```

## Problem 4: Cab Fare Estimator with Retry Option

### Rules:
- Base fare = ₹50
- ₹12 per km
- Peak hour → 25% extra
- Allow retry using while loop

### Real-Time Use:
- Ride-hailing apps
- Transport systems


```python
# Problem 4: Cab Fare Estimator

def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (distance * 12)

    if peak.lower() == "yes":
        fare += fare * 0.25

    return fare


while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak_hour = input("Is it peak hour? (yes/no): ")

        total_fare = calculate_fare(distance, peak_hour)
        print(f"Total Fare: ₹{total_fare}")

        break
    except ValueError:
        print("Please enter valid input.")
```

## Problem 5: Employee Attendance Eligibility Checker

### Rules:
- Calculate attendance percentage
- Eligible if ≥ 75%
- Not Eligible otherwise

### Real-Time Use:
- HR payroll systems
- Internship tracking


```python
# Problem 5: Employee Attendance Eligibility Checker

def check_attendance(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    for status in attendance_list:
        if status == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100

    print(f"Attendance Percentage: {percentage}%")

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

eligibility = check_attendance(attendance)
print("Status:", eligibility)
```

## Problem 6: Password Strength Checker

### Rules:
- Minimum length = 8
- Must contain at least one digit
- Must contain at least one special character (@ # $)

### Real-Time Use:
- Login systems
- Security validation


```python
# Problem 6: Password Strength Checker

def check_password(password):
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Password too short"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


user_password = input("Enter password: ")
print(check_password(user_password))
```
